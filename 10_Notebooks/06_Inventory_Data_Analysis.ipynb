```python
import pandas as py
inventory =py.read_csv(r"C:\Users\ADMIN\Desktop\Retail_Analytics_Project\02_Raw_Data\Inventory_Data.csv")
print(inventory.head().to_string())
```

```python
print(" information of inventory")
print(inventory.info())

print("shape of the inventory")
print(inventory.shape)

print("describe the data set")
print(inventory.describe())

print(" checking null values")
print(inventory.isnull().sum())
print(inventory["Site ID"].isnull().sum())
print(inventory["Product ID"].isnull().sum())

print("unique value is site id, pro id")
print(inventory["Site ID"].nunique())
print(inventory["Product ID"].nunique())

print("checking Stockout Flag")
print(inventory.value_counts().sum())
```

```python
inventory["Net Stock"]=inventory["Ending Inventory"]-inventory["Beginning Inventory"]
print(f"{inventory.head()}")
```

```python
inventory["Stockout Flag"].value_counts()
```

```python
print(inventory.duplicated().sum())
```

```python
print("checking the negative value")
print((inventory["Beginning Inventory"]< 0).sum())
print((inventory["Ending Inventory"]< 0).sum())
print((inventory["Replenishment"]< 0).sum())
```

```python
inventory.sort_values(
    "Net Stock"
).head()
```

```python
inventory[~inventory["Stockout Flag"].isin(["Yes","No"])]
```

```python
text_column=["Site ID","Product ID","Stockout Flag"]
for col in text_column:
    inventory[col]=inventory[col].str.strip()
print(inventory.head())
```

```python
inventory.to_csv(
"../03_Cleaned_Data/Inventory_Data_Cleaned.csv",
index=False
)
```
