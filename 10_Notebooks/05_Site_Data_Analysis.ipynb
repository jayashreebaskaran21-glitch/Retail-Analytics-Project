```python
import pandas as py
sites=py.read_csv(r"C:\Users\ADMIN\Desktop\Retail_Analytics_Project\02_Raw_Data\Site_Details.csv")
print(sites.head().to_string())
```

```python
# info
print(sites.info())
# describe
print(sites.describe())
# shape 
print( sites.shape)
# columns
print("columns")
print(sites.columns.tolist())

#isnull()
print(sites.isnull().sum())

```

```python
# unquie
print(sites["Site ID"].nunique())

print(sites["Site ID"].is_unique)
print(sites["Site ID"].duplicated().sum())

```

```python
# value of counts
print(sites["City"].value_counts())
print(sites["Region"].value_counts())
print(sites["State"].value_counts())
print(sites["Status"].value_counts())
print(sites["Site Format"].value_counts())
```

```python
# str to time
sites["Open Date"] = py.to_datetime(sites["Open Date"],format="%d-%m-%Y")
print(sites.info())
```

```python
# str to time
sites["Open Date"] = py.to_datetime(sites["Open Date"],format="%d-%m-%Y")
print(sites.info())
```

```python


# alignment in column

sites.columns=sites.columns.str.strip()

print(sites.head())
```

```python
#isna
print(sites.isna().sum())
```

```python
#1. Store Size should be positive
print((sites["Store Size"]<=0).sum())
```

```python
#2. Open Date should not be in the future
print((sites["Open Date"]> py.Timestamp.today()).sum())
```

```python
print(sites["Status"].nunique())
print(sites[~sites["Status"].isin(["Active","Inactive"])].value_counts())
```

```python
text_cols = [
    "Site ID",
    "Site Name",
    "Site Format",
    "Region",
    "City",
    "State",
    "Status"
]

for col in text_cols:
    sites[col] = sites[col].str.strip()
```
