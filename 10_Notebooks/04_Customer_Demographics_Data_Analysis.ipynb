```python
import pandas as py
customer_demographic=py.read_csv("../02_Raw_Data/Customer_Demographics.csv")
print(customer_demographic.head().to_string())
```

```python
#customer_demographic
#1 shape
print(customer_demographic.shape)

#2 describe
print(customer_demographic.describe())

#3 cloumns

print(customer_demographic.columns)

#4 info

print(customer_demographic.info())

#5 duplicate 


print(customer_demographic.duplicated().sum())

```

```python
#6 unquie
print(customer_demographic["Customer ID"].nunique())

#7 isnull
print(customer_demographic[["Customer ID","Age","Gender","Income Bracket","Purchase Frequency","Average Spend"]].isnull().sum()
)
#8 TO LIST
print(customer_demographic.columns.tolist())
# 9 IS NULL SUM
print(customer_demographic.isnull().sum())
```

```python
# values of counts

print(customer_demographic["Gender"].value_counts())

# remove extra space
customer_demographic.columns = customer_demographic.columns.str.strip()
print(customer_demographic.columns)

# validation income bracke
print(customer_demographic["Income Bracket"].value_counts())
print((customer_demographic["Purchase Frequency"] <= 0).sum() )
```

```python
print((customer_demographic["Average Spend"] <= 0).sum() )
print((customer_demographic["Average Spend"] <= 0).all () )
print((customer_demographic["Average Spend"] <= 0).any () )
```

```python
customer_demographic[(customer_demographic["Age"])<18| (customer_demographic["Age"]>100)]
```

```python
#Customer ID Duplicate Check

customer_demographic[
    customer_demographic["Customer ID"].duplicated(keep=False)
]
```

```python
#Gender Validation
customer_demographic[customer_demographic["Gender"].isin(["Male","Female","Other"])]
```

```python
gen=["Male","Female","Other"]
(customer_demographic["Gender"].isin(gen)).sum()
```

```python
valid_income = [
    "1-5 LPA",
    "10 LPA",
    "10-20 LPA",
    "20+ LPA"
]

customer_demographic[
    ~customer_demographic["Income Bracket"].isin(valid_income)
]
```

```python
customer_demographic.sort_values(
    "Average Spend",
    ascending=False
).head(10)
```

```python
#Remove Spaces from Text Columns

text_cols = [
    "Customer ID",
    "Gender",
    "Income Bracket"
]

for col in text_cols:
    customer_demographic[col] = customer_demographic[col].str.strip()
```

```python
print(customer_demographic.info())

print(customer_demographic.describe())

print(customer_demographic.head())
```

```python
customer_demographic.to_csv("../03_Cleaned_Data/Customer_Demographics_Cleaned.csv")
```
