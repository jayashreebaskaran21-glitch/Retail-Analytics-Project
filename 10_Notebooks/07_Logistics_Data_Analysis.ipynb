```python
import pandas as py
logistics =py.read_csv(r"C:\Users\ADMIN\Desktop\Retail_Analytics_Project\02_Raw_Data\Logistics_Data.csv")
print(logistics.head().to_string())
```

```python
print(" logistics shape")

print(logistics.shape)

print("logistics information")

print(logistics.info())

print("logistics describe")

print(logistics.describe())

print("isnull value")

print(logistics.isnull().sum())
print(logistics[["Shipment ID","Site ID","Product ID"]].isnull().sum())

print("Unique value")
print(logistics[["Shipment ID","Site ID","Product ID"]].nunique().sum())

print(" duplicate value")
print(logistics[["Shipment ID","Site ID","Product ID"]].duplicated().sum())
```

```python
print("QUANTITY VALUE CHECKING")

print((logistics["Quantity"]<=0).sum())
```

```python
#logistics["Shipment Date"]=py.to_datetime(logistics["Shipment Date"],format="%d-%m-%y")
logistics["Shipment Date"] = py.to_datetime(
    logistics["Shipment Date"],
    errors="coerce"
)
```

```python
print(logistics.head())
print(logistics.info())
```

```python
print(" VALUE COUNT")
print(logistics["Transportation Type"].value_counts())
print(logistics["Delivery Status"].value_counts())
```

```python
trans=["Ship","Rail","Air","Truck"]
print(logistics["Transportation Type"].isin(trans).value_counts().sum())
```

```python
Delivery=["Delivered","Delayed","Cancelled"]
print(logistics["Delivery Status"].isin(Delivery).sum())
```

```python
logistics.columns.to_list()
```

```python
text_clo=["Shipment ID", "Site ID", "Product ID","Delivery Status", "Transportation Type"]
for col in text_clo:
    logistics[col]=logistics[col].str.strip()
```

```python
print(logistics[logistics["Shipment Date"].isna()])
```

```python
valid_status = [
    "Delivered",
    "Delayed",
    "Cancelled"
]

print(
    logistics[
        ~logistics["Delivery Status"].isin(valid_status)
    ]
)
```
