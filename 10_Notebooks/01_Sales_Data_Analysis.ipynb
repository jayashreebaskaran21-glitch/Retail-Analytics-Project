```python
total_revenue=sales_clean["Revenue"].sum()
print(f"Total Revenue: {total_revenue: ,.2f}")

total_sold=sales_clean["Units Sold"].sum()
print(f"Total Unit Sold: {total_sold: ,.2f}")

total_discount = sales_clean["Discounts"].sum()

print(f"Total Discounts : ₹{total_discount:,.2f}")

total_return = sales_clean["Returns"].sum()

print(f"Total Returns : ₹{total_return:,.2f}")


```

```python
top_10=sales_clean.groupby("Product ID")["Revenue"].sum().sort_values(ascending=False)
bottom=sales_clean.groupby("Product ID")["Revenue"].sum().sort_values()

revenue =(sales_clean.groupby("Product ID",as_index=False)["Revenue"].sum().sort_values("Revenue",ascending=False))
unitd =sales_clean.groupby("Product ID",as_index=False)["Units Sold"].sum().sort_values("Units Sold",ascending=False)
print(f"Top Product -\n{top_10.head()}")
print(f"Least Product -\n{bottom.head()}")
```

```python
print(f"Revenue -\n{revenue.head(10)}")
print(f"Unit Sold-\n{unitd.head(10)}")
```

```python
revenue_by_product = (
    sales_clean
    .groupby("Product ID", as_index=False)["Revenue"]
    .sum()
    .sort_values("Revenue", ascending=False)
)

revenue_by_product.head()
```

```python
# STORE ANALYSIS BY REVENUE BY STORE,UNIT SOLD BY STORE,RETURN RATE BY STORE
revenue_store=sales_clean.groupby("Site ID",as_index=False)["Revenue"].sum().sort_values("Revenue",ascending=False)
print(f"Revenue by Store\n{revenue_store.head()}")

unit_store=sales_clean.groupby("Site ID",as_index=False)["Units Sold"].sum().sort_values("Units Sold",ascending=False)
print(f"Unit Sold by Store\n{unit_store.head()}")

return_store=sales_clean.groupby("Site ID",as_index=False)["Returns"].sum().sort_values("Returns",ascending=False)
print(f"Revenue by Store\n{return_store.head()}")
```

```python
store_analysis = (
    sales_clean
    .groupby("Site ID", as_index=False)[["Units Sold", "Returns", "Revenue"]]
    .sum()
)
store_analysis["Return Rate (%)"] = (
    store_analysis["Returns"] /
    store_analysis["Units Sold"]
) * 100


store_analysis.head().sort_values("Revenue", ascending=False)
```

```python
store_summary = (
    sales_clean
    .groupby("Site ID", as_index=False)
    .agg({
        "Revenue": "sum",
        "Units Sold": "sum",
        "Returns": "sum"
    })
)

store_summary["Return Rate (%)"] = (
    store_summary["Returns"] /
    store_summary["Units Sold"]
) * 100

store_summary = store_summary.sort_values("Revenue", ascending=False)

store_summary.head()
```

```python
sales_clean["Month"] = sales_clean["Date"].dt.month_name()
sales_clean["Month_No"] = sales_clean["Date"].dt.month
```

```python


units_month = (
    sales_clean
    .groupby(["Month_No", "Month"], as_index=False)["Units Sold"]
    .sum()
    .sort_values("Month_No")
)

discount_month = (
    sales_clean
    .groupby(["Month_No", "Month"], as_index=False)["Discounts"]
    .sum()
    .sort_values("Month_No")
)

returns_month = (
    sales_clean
    .groupby(["Month_No", "Month"], as_index=False)["Returns"]
    .sum()
    .sort_values("Month_No")
)
```

```python
print("Revenue by Month")
print(revenue_month)

print("\nUnits Sold by Month")
print(units_month)

print("\nDiscount by Monthr")
print(discount_month)

print("\nReturns by Month")
print(returns_month)
```
